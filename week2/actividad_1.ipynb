```python
import numpy as np
def step(z):
    return 1 if z >= 0 else 0
def neurona(x1, x2, w1, w2, b):
    z = x1*w1 + x2*w2 + b
    y = step(z)
    return z, y
# Valores base
w1, w2, b = 1, 1, -1.5 # intenta parecerse a una compuerta AND
casos = [(0,0),(0,1),(1,0),(1,1)]
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2, b)
    print(f"Entrada=({x1},{x2}) z={z:.1f} salida={y}")
```

```python
# Test with the current b value (which is correct for AND gate)
print("\n--- Testing with b = -1.5 (AND gate behavior) ---")
w1, w2, b_and = 1, 1, -1.5
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2, b_and)
    print(f"Entrada=({x1},{x2}) z={z:.1f} salida={y}")

# Test with a b value outside the range (e.g., b = -0.5), which would act like an OR gate
print("\n--- Testing with b = -0.5 (OR gate behavior) ---")
w1, w2, b_or = 1, 1, -0.5
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2, b_or)
    print(f"Entrada=({x1},{x2}) z={z:.1f} salida={y}")

# Test with another b value outside the range (e.g., b = -2.5), which would output 0 for (1,1)
print("\n--- Testing with b = -2.5 (Always 0 for (1,1)) ---")
w1, w2, b_fail = 1, 1, -2.5
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2, b_fail)
    print(f"Entrada=({x1},{x2}) z={z:.1f} salida={y}")
```
