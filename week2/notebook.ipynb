```python
import numpy as np
import pandas as pd

# -----------------------------
# 1. Definir pesos y bias
# -----------------------------
# Puedes cambiarlos para ver cómo afecta la decisión
w = np.array([1.0, -1.0])   # pesos w1, w2
b = 0.5                     # bias

print("Pesos w:", w)
print("Bias b:", b)

# -----------------------------
# 2. Definir la neurona (perceptrón)
# -----------------------------
def perceptron(x):
    """
    x: vector de entrada, por ejemplo np.array([x1, x2])
    Retorna:
        z  -> puntaje (suma ponderada + bias)
        y -> salida binaria (0 o 1)
    """
    # Cálculo del puntaje z = w·x + b
    z = np.dot(w, x) + b
    # Regla de decisión (función escalón)
    y = 1 if z >= 0 else 0
    return z, y

# -----------------------------
# 3. Conjunto de pruebas (casos controlados)
# -----------------------------
# Cada fila es un par [x1, x2]
X_test = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1],
    [2, -1],
    [-1, 2]
])

# -----------------------------
# 4. Cálculo de z y salida para cada ejemplo
# -----------------------------
resultados = []
print("\nResultados por caso de prueba:")
print("x1\t x2\t z\t\t salida")
for x in X_test:
    z, y = perceptron(x)
    resultados.append({
        "x1": x[0],
        "x2": x[1],
        "z": z,
        "salida": y
    })
    print(f"{x[0]:.2f}\t {x[1]:.2f}\t {z:.4f}\t {y}")

# -----------------------------
# 5. Tabla/resumen en DataFrame
# -----------------------------
df_resultados = pd.DataFrame(resultados)
print("\nTabla resumen:")
display(df_resultados)
```

## Análisis del perceptrón: efecto de pesos y bias

En esta actividad se implementó una neurona tipo perceptrón con dos entradas \(x_1, x_2\), pesos \(w_1, w_2\) y un término de bias \(b\). El puntaje \(z\) se calculó como una combinación lineal de las entradas más el bias, es decir \(z = w_1 x_1 + w_2 x_2 + b\), y la decisión de salida se tomó aplicando una función de activación escalón: si \(z \ge 0\) la salida es 1 y si \(z < 0\) la salida es 0.

Al observar la tabla de resultados generada en las pruebas, se puede ver que cambios en las entradas modifican el valor de \(z\) y, cuando el signo de \(z\) cambia, también cambia la clase asignada. Los pesos controlan qué tan importante es cada entrada: si un peso es grande en valor absoluto, pequeñas variaciones de esa entrada producen cambios grandes en \(z\), afectando la clasificación final.

El bias desplaza el umbral de decisión. Cuando el bias \(b\) aumenta, el valor de \(z\) tiende a ser más positivo para las mismas entradas, por lo que más casos se clasifican como 1. Si el bias se hace más negativo, ocurre lo contrario: la mayoría de los puntajes \(z\) quedan por debajo de cero y el perceptrón tiende a predecir 0 en más ejemplos.

En conclusión, los pesos determinan la orientación de la frontera de decisión en el espacio de características, mientras que el bias determina su desplazamiento. Ajustar estos parámetros permite mover la frontera de decisión y cambiar qué puntos quedan en la clase 1 y cuáles en la clase 0, lo que explica el comportamiento observado en las salidas del modelo.
