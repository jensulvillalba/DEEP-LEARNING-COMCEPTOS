## 1. Importar librerías necesarias

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Crear un dataset simple de clasificación binaria
 ejemplo similar a una compuerta lógica.

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
## Aquí se define el conjunto de datos de entrada X y las etiquetas y.
y = np.array([0,0,0,1])  # Compuerta AND
```

## 3. Implementar el Perceptrón

```python
class Perceptron: ##se define el perceptron

    def __init__(self, num_features, learning_rate=0.01, epochs=20): ##Se definen los parámetros del modelo:
        self.weights = np.zeros(num_features + 1)  # incluye bias
        self.lr = learning_rate  ##qué tan rápido se actualizan los pesos
        self.epochs = epochs     ##número de iteraciones de entrenamiento
        self.errors = []

    def activation_function(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):   ## funcion de prediccion
        z = np.dot(x, self.weights[1:]) + self.weights[0]
        return self.activation_function(z)

    def train(self, X, y): ##Esta función se encarga de entrenar el perceptrón
        for epoch in range(self.epochs):
            total_error = 0

            for xi, target in zip(X, y):
                prediction = self.predict(xi)
                update = self.lr * (target - prediction)

                self.weights[1:] += update * xi
                self.weights[0] += update

                total_error += abs(target - prediction)

            self.errors.append(total_error)
```

## 4. Entrenar el modelo

```python
perceptron = Perceptron(num_features=2)
perceptron.train(X, y)

print('Pesos finales:', perceptron.weights)
```

## 5. Probar el perceptrón

```python
for xi in X:  ## esta funcion muestra si el modelo aprendió correctamente.
    print('Entrada:', xi, 'Predicción:', perceptron.predict(xi))
```

## 6. Visualizar el error durante el entrenamiento

```python
plt.plot(perceptron.errors)
plt.title('Error por época durante el entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Error')
plt.show()
```

## 7. Conclusiones
- El perceptrón ajusta sus pesos mediante un proceso iterativo.
- La función de activación permite realizar clasificación binaria.
- A medida que avanza el entrenamiento el error disminuye.
- Esto demuestra el proceso básico de aprendizaje en redes neuronales.
